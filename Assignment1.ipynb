importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the given data & printing it.

```python
df = pd.read_csv('Auto.csv')
df
```

(a)- It's clear by looking at data that name is a qualitative predictor and all others are quantitative predictors.

Now, we shall check if our data has any missing values.

```python
column_names = df.columns
count=0
for col in column_names:
  if df[col].isnull().sum()!=0:
    count=1
  else:
    print(col, ' checked. No null values found')
print(count)
```

```python
for col in column_names:
  string=0
  for value in df[col].values:
    if str(type(value))== "<class 'str'>":
      string=1
  if string==1:
    print(col, ' either has datatype string or has missing values')
```

```python
sum=0
int_count=0
for value in df['horsepower'].values:
  if value!='?':
    int_count+=1
    sum+= int(value)
mean= int(sum/int_count)
df['horsepower']= df['horsepower'].replace('?',mean)
df.head(40)
df['horsepower']= df['horsepower'].astype(int)
```

```python
#b
for col in column_names:
    print('Values of ',col, ' ranges from ', np.min(df[col]), ' to ', np.max(df[col]))
```

```python
#c
for col in column_names:
  if col!= 'name':
    mean= np.mean(df[col])
    std= np.std(df[col])
    print(col, ' has mean = ', mean, ' and standard deviation= ', std)
```

```python
#d
df= df.drop(df.index[10:85])
df= df.reset_index(drop= True)
df.head(40)
```

Now we removed the data from 10th to 85th observation using drop function. Then, we reset the index to make data appear more presentable.
Now let's calculate the range, mean and standard deviation of remaining data.

```python
#d
for col in column_names:
  if col!= 'name':
    mean= np.mean(df[col])
    std= np.std(df[col])
    print('Values of ',col, ' ranges from ', np.min(df[col]), ' to ', np.max(df[col]))
    print(col, ' has mean = ', mean, ' and standard deviation= ', std)
```
